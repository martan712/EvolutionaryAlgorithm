```python
def write_output(fitnesses, filename):
    import csv

    with open(f"{filename}.csv", "w", newline='') as f:  # FIX: Set newline=''
        wr = csv.writer(f)
        wr.writerows(fitnesses)
```

```python
from GeneticAlgorithm import GeneticAlgorithm, generate_random_string
import string
fitnesses = []
# Example usage
alphabet = string.ascii_letters
L = 15
target = generate_random_string(alphabet, L)  # Target string
for x in range(10):
    ga = GeneticAlgorithm(alphabet, target, L, mu=1/L, K=2)
    fitness, distribution = ga.run()
    fitnesses.append(fitness)
write_output(fitnesses, "b1.1")
```
