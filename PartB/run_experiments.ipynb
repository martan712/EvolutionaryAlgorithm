```python
def write_output(fitnesses, filename):
    import csv

    with open(f"{filename}.csv", "w", newline='') as f:  # FIX: Set newline=''
        wr = csv.writer(f)
        wr.writerows(fitnesses)
```

```python
ga = GeneticAlgorithm(alphabet, target, L, mu=1/L, K=2)
fitness, distribution = ga.run()

ga = GeneticAlgorithm(alphabet, target, L, mu=0, K=2)
fitness2, distribution2 = ga.run()

ga = GeneticAlgorithm(alphabet, target, L, mu=3/L, K=2)
fitness3, distribution3 = ga.run()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate three distributions
np.random.seed(42)
dist1 = distribution
dist2 = distribution2
dist3 = distribution3

# Combine into a single dataset
data = {
    "Value": np.concatenate([dist1, dist2, dist3]),
    "Group": ["A"] * 200 + ["B"] * 200 + ["C"] * 200  # Labels for each distribution
}

# Create the beeswarm plot
plt.figure(figsize=(8, 5))
sns.swarmplot(x=data["Group"], y=data["Value"], size=5, palette="viridis")

# Labels and title
plt.xlabel("Distribution Group")
plt.ylabel("Value")
plt.title("Beeswarm Plot of Three Distributions")

# Show the plot
plt.show()
```
